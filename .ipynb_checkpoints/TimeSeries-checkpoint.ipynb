## Stationarity

모델링하는 데 필수<br>
trend와 seasonality없는 상태<br>

- 검사방법<br>
Augmented DickeyFuller test<br>
H0(nonstationary) p값 작을 수록 stationary하다는 뜻<br>

원상태로는 설명이 없어서 직접 달아줘야하는 게 귀찮음<br>
AIC(Akaike information criterion), BIC(Bayesian IC) 낮은 게 좋은 모델 평가기준(tool에 내장되어 혼자쓰일때는 드물다)<br>

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series):
    result = adfuller(sales, autolag='AIC')
    labels = ['ADF test statistic','p-value','# of lags used', '# of observations']
    out = pd.Series(result[0:4], index=labels)
    for key, val in result[4].items():
        out[f'critical value({key})'] = val
adf_test(sales)
```

- Differencing으로 stationary하게 만들 수 있음<br>
무조건 예측 후에 되돌려 놔야함<br>

trend가 linear하지 않다면 second order사용해야 할 수도 있음(단순함)<br>

```python
sales['dif1'] = sales.diff()  # 단순하게 전 시점에을 빼준 것이기에 내장
sales['dif2'] = sales.diff().diff() 
```

```python
sales['forecast'] = sales.iloc[-1] + sales['dif1forecast'].cumsum()  # 진짜 데이터의 마지막 값에 예측의 누적합 series를 더해 복원
```

```python
sales['dif1forecast'] = (sales.iloc[-1] - sales.iloc[-2]) + sales['dif2forecast'].cumsum()
sales['forecast'] = sales.iloc[-1] + sales['dif1forecast'].cumsum
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

train_df, test_df = sales.iloc[:2500000], sales.iloc[2500000:]
sales['Simple'] = sales.ewm(span=12).mean()  #default로 (1-alpha)^i,alpha=2/(span+1) #adjust=False하면 복잡

double_model = ExponentialSmoothing(train_df, trend='add').fit()
triple_model = ExponentialSmoothing(train_df, trend='add', seasonal='add', seasonal_periods=360).fit()  #season은 몇 freq인가 #판단이 더 어려워서 metric으로
sales['Double'] = double_model.fittedvalues
sales['Triple'] = triple_model.fittedvalues
print(sales)
```

## Cross validation

시계열 데이터는 전후 데이터간 상관관계가 존재하기에 기존 cv 방법 활용 불가(시간순으로 나열된 데이터 보존)<br>
훈련데이터는 항상 검증데이터에 선행해야함<br>
훈련데이터를 점점 늘려나가거나(크기증가) 다른 시점으로 잡거나(크기고정)<br>

두번째 방법은 해결하기 위해 훈련 검증 데이터간 공백두고 fold간 공백두고<br>

```python
class BlockingTimeSeriesSplit():
    def __init__(self, n_splits): 
        self.n_splits = n_splits
    def split(self, X):
        n_samples = len(X)
        k_fold_size = n_samples // self.n_splits
        indices = np.arange(n_samples)
        margin = 0  # margin도 두어야함
        for i in range(self.n_splits):
            start = i * k_fold_size
            stop = start + k_fold_size
            mid = int(0.8 * (stop - start)) + start  #각 fold별 test를 0.2
            yield indices[start : mid], indices[mid+margin : stop]
```
