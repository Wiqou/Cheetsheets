# Get 요청

```python
import requests
res = requests.get('http://ysweb.yonsei.ac.kr:8888/curri120601/curri_pop_mileage_result01.jsp').text
```

# BeautifulSoup

### find(), find_all()
태그 검색   
- find(tag, attributes, recursive, text, keywords)   
- find_all(tag, attributes, recursive, text, limit, keywords)   

### get_text()
태그 내 텍스트 반환(부모 태그의 경우 자식 태그의 텍스트를 모두 반환)

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res, 'lxml') 
tag = soup.find('h3')  # id나 class도 인수로 검색가능
print(tag)
tag.get_text()
```

# Selenium

동적페이지에서 크롤링하기 위한 도구(대신 좀 느림)

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
driver = webdriver.Chrome("C:\\Users\\Dongjun\\chromedriver_win32\\chromedriver.exe")
driver.get('https://www.naver.com/')

search = driver.find_element_by_id('query')  # id 속성으로 검색창 찾기
search.send_keys('제주도')  # 검색어 입력
search.send_keys(Keys.RETURN)  # Keys.RETURN = Enter 입력
search = driver.find_element_by_link_text('이미지')  # a 태그의 텍스트로 찾기
search.click()  # 클릭

driver.close
```

셀레니움 사용 시 브라우저에서 해당 웹 페이지의 요소들을 로드하는 데 시간이 좀 걸리기 때문에, element가 존재하지 않는다는 error가 발생할 수 있다.   
그럴 경우 해당 요소가 전부 준비가 될 때까지 대기해야 한다.

driver.implicitly_wait(time_to_wait=5)   # 위 방법은 암시적 대기 방식으로, webdriver에 영구적으로 작용된다.
